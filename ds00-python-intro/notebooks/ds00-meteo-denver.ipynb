### Task 1

```python
# import libraries
import matplotlib.pyplot as plt  
import numpy as np

# load data, sort by \t
tmax = np.loadtxt(fname='../data/meteo_denver_tmax_2000_2022.txt', delimiter='\t')

# create an array for the calendar months and for years 2000-2022
months = np.arange(1, 13)
years = np.arange(2000, 2023)

# generate the map and add an appropriate color scale
fig, ax = plt.subplots(1,1, figsize=(5,5))   
cs = plt.pcolormesh(months,years,tmax, cmap='hot') 

# label the axes and add title
ax.set_xlabel('Month', fontsize=14) 
ax.set_ylabel('Year', fontsize=14)
cbar = plt.colorbar(cs)
cbar.set_label('Maximum Temperature [°F]', fontsize=14, labelpad=14)
plt.title('Max Temperature at Denver Water Dept.', fontsize=18 )
```

#### You can see that June-August are the hottest months over the 22-year period. Conversely, December, January, and February trend as the coldest months. January 2007 was an especially cold month. 

### Task 2

```python
# create subplots to create two histograms to visualize bins
fig, axes = plt.subplots(1,2, figsize=(10,5)) 

# plot both histograms
axes[0].hist(tmax.flatten(), bins=30, color= '#F08000', ec = '#6E260E')
axes[1].hist(tmax.flatten(), bins=15, color= '#F08000', ec = '#6E260E')

# add titles and axes labels
axes[0].set_title('Histogram with 30 Bins') 
axes[1].set_title('Histogram with 15 Bins')
axes[0].set_xlabel('Maximum Temperature [°F]')
axes[1].set_xlabel('Maximum Temperature [°F]')
axes[0].set_ylabel('Number of Occurences')
axes[1].set_ylabel('Number of Occurences')
plt.suptitle('Max Temperature at Denver Water Dept.')

```

#### These histograms show us the amount of days per year in a given temperature range. Changing the number of bins changes the range of each bin. The total temperature / number of bins = the temperature range of each bin. Having more bins gives a more detailed view of the data, and less bines give a better overview but may cause misinterpretation of mean or mode. 

### Task 3

```python
# assign variables to the monthly mean, the standard deviation, and the upper/lower bound
monthly_mean_max = np.mean(tmax, axis=0)
standard_dev = np.std(tmax, axis=0, ddof=0)
upper_bound = monthly_mean_max + standard_dev
lower_bound = monthly_mean_max - standard_dev

# generate a plot of the results
plt.figure(figsize=(10, 5))
plt.plot(months, monthly_mean_max, label="Average Max Temperature", color='blue', marker='o')
plt.plot(months, upper_bound, label="Avg + Std Dev", color='red', linestyle='dashed', marker='^')
plt.plot(months, lower_bound, label="Avg - Std Dev", color='green', linestyle='dashed', marker='v')

# create axes labels and title
plt.xlabel("Month")
plt.ylabel("Temperature (°F)")
plt.title("Monthly Average Max Temperature at Denver Water Dept. with Standard Deviation")
plt.legend()
plt.grid(True)
```

#### This plot is very similar to the columns of the heat map in Task 1, as it also shows July and August as the hottest months, and February and January as the coldest. One difference is that it only shows averages, so we cannot specifically look at the months in each individual year.

### Task 4

```python
#assign variable to yearly average, use numpy to generate the average per year
yearly_avg = np.mean(tmax, axis=1)

# define x-axis (years)
years = np.arange(1, len(yearly_avg) + 1)

# generate the figure
plt.figure(figsize=(12, 6))
plt.plot(years, yearly_avg, marker='o', color='salmon', linestyle='-')

#label axes and add title
plt.xlabel("Year (2000-2022)", size=12)
plt.ylabel("Average Max Temperature (°F)", size=12)
plt.title("Yearly Average Maximum Temperature at Denver Water Dept.", size=16)
plt.xticks(np.arange(0.5, 23), np.arange(2000, 2023), rotation=90, size=8)
plt.grid(True)
```

#### This plot is similar to the rows of the plot from Task 1. 2012 had the warmest maximum temperature, while 2007 had the lowest maximum temperature, on average.
